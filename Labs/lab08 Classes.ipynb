```python
import time

class Animal:
    
    animalNumber = 0
    
    # __init__ runs when an Animal is constructed
    # Length is in feet
    def __init__(self, species, name):
        Animal.animalNumber += 1
        self.id = Animal.animalNumber
        self.species = species
        self.name = name
        self.last_feeding = False
        
    def display(self):
        print("\n\tSubject: %s" % self.name)
        print("\tID: %d" % self.id)
        print("\tSpecies: %s" % self.species)
        if self.last_feeding:
            print("\tLast Feeding: %s" % self.last_feeding)
        else:
            print("\tNever Fed :(")
        
    def feed(self):
        self.last_feeding = time.ctime()
```

```python
class Felis_catus(Animal):
    def __init__(self, species, name, happiness, breed, jerklevel):
        self.happiness = happiness
        self.breed = breed
        Animal.__init__(self, species, name)
    
    def pet(self, length):
        self.happiness += length - jerklevel
    
    def laser(self, length):
        self.happiness += (length * 4) - jerklevel
        
    def wait(self, length):
        self.happiness -= (length/2) - jerklevel
    
    def display(self):
        print("%s is %d happy!" % (self.name, self.happiness))
        
waffles = Felis_catus('catus', 'Waffles', 3, 'Calico', 0)
chad = Felis_catus('catus', 'Chad', 2, 'Tabby', -2)
ambuck = Felis_catus('catus', 'Ambuck', 5, 'Ragekitty', -3)
action = ""
        
while True:
    yescat = input("Would you like to pet a cat or play with a laser?")
    if yescat == "no" :
        print("Agreed. That could be a dangerous endevor.")
        break
    if yescat == "yes" :
        cat = input("Which cat would you like?")
        action = (input("which would you like to do?"))
        if action == "pet" :
            length = int(input("How long?"))
            if cat == "Waffles" :
                waffles.pet(length)
                chad.wait(length)
                ambuck.wait(length)
            elif cat == "Chad" :
                chad.pet(length)
                waffles.wait(length)
                ambuck.wait(length)
            elif cat == "Ambuck" :
                ambuck.pet(length)
                waffles.wait(length)
                chad.wait(length)
            else :
                print("pffff!! That's not a cat!")
                continue
        elif action == "laser" :
                length = int(input("How long?"))
                if cat == "Waffles" :
                    waffles.pet(length)
                    chad.wait(length)
                    ambuck.wait(length)
                elif cat == "Chad" :
                    chad.pet(length)
                    waffles.wait(length)
                    ambuck.wait(length)
                elif cat == "Ambuck" :
                    ambuck.pet(length)
                    waffles.wait(length)
                    chad.wait(length)
                else :
                    print("pffff!! That's not a cat!")
                    continue
        
    waffles.display()
    chad.display()
    ambuck.display()
    if waffles.happiness < 0 :
        print("Waffles pushes your glass cup to the floor with a crash! If he wasn't so coot...")
    if chad.happiness < 0 :
        print("Chad notices your lack of attention and begins scratching the couch. !@% Chad!")
    if ambuck.happiness < 0 :
        print("Ambuck is not amused by your independence, Ambuck gives you a look of indiference.")
```

```python
# getAnimal will ask user for animal information and return an Animal object
def getAnimal():
    name = input("\n\tName: ")
    species = input("\tSpecies: ")
    animal = Animal(species, name)
    print("\tAnimal Created, ID: %d" % animal.id)
    return animal

registry = {}

# main program loop
while True:
    cmd = input("(A)dd (R)emove (L)ookup (F)eed (Q)uit")
    if cmd == "Q":
        print("Goodbye!")
        break
    if cmd == "A":
        animal = getAnimal()
        registry[animal.id] = animal
    if cmd == "L":
        id = int(input("\tInput ID: "))
        if not id in registry:
            print("\tAnimal ID %d Not Found!" % id)
            continue
        registry[id].display()
    if cmd == "R":
        id = int(input("\tInput ID: "))
        del registry[id]
        print("\tAnimal ID %d Deleted!" % id)
    if cmd == "F":
        id = int(input("\tInput ID: "))
        registry[id].feed()
        print("\tAnimal ID %d Fed!" % id)
```
