```python
from ipyturtle import Turtle
t = Turtle(fixed=False, width=320, height=320)

def circle():
        for i in range(360):
            t.forward(1)
            t.left(1)
            
def square():
    t.forward(60)
    t.left(90)
    t.forward(120)
    t.left(90)
    t.forward(120)
    t.left(90)
    t.forward(120)
    t.left(90)
    t.forward(60)
    
def triangle():
    t.left(30)
    t.forward(120)
    t.left(120)
    t.forward(120)
    t.left(121)
    t.forward(120)
    t.left(90)

def trapazoid():
    t.left(30)
    t.forward(60)
    t.left(60)
    t.forward(60)
    t.left(60)
    t.forward(60)
    t.left(121)
    t.forward(120)
    t.left(90)
    
def qturn():
    t.penup()
    t.right(90)
    t.forward(5)
    t.left(90)
    t.forward(5)
    t.left(15)
    t.pendown()
    
t.penup()
t
```

```python
shape = int(-1)
while True:
    print("\n\nPlease pick a shape:")
    print("(1) Circle, (2) Square, (3) Triangle, (4) Trapazoid")
    shape = int(input("What's your shape? "))
    index = int(input("How many do you want?")) + 1
    t.reset()
    if shape < 2:
        while index > 0:
            circle()
            qturn()
            index += -1
        print("That's super great, right?!")

    elif shape < 3:
        while index > 0:
            square()
            qturn()
            index += -1
        print("Pretty lit square you got there!")
    
    elif shape < 4:
        while index > 0:
            triangle()
            qturn()
            index += -1
        print("That's isocoles!")
    
    elif shape < 5:
        while index > 0:
            trapazoid()
            qturn()
            index += -1
        print("That's mathmatical!")
    else:
        break 
```
