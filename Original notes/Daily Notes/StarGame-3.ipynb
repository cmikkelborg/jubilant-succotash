```python
%%python

from tkinter import *
import time

# make the tk window
tk = Tk()

# set a title
tk.title("Cardassian Blast")

# make the window resizable
tk.resizable(0, 0)

# put the window on top
tk.wm_attributes("-topmost", 1)

# create a canvas
canvas = Canvas(tk, width=900, height=700, bd=0, bg="black", highlightthickness=0)

# put the canvas in the tk window
canvas.pack()

# initialize our window
tk.update()

def fireButton():
    global phaserCount
    phaserCount = 10
    
# ship class
class Ship:
    def __init__(self, canvas, image, x, y):
        self.canvas = canvas
        self.id = canvas.create_image(x, y, anchor=NW, image=image)
    def fly_NE(self):
        self.canvas.move(self.id, 1, -1)
    def fly_E(self):
        self.canvas.move(self.id, .5, 0)
    def firePhaser(self, x, y, angle):
        #print("Firing Phaser from %s,%s at angle %s" % (x, y, angle))
        lineID = self.canvas.create_line(x, y, x+500, y+500, fill="red", width=7)
        return lineID
    
btn = Button(tk, text="FIRE!", command=fireButton)
btn.pack()

# falcon
falcon_image = PhotoImage(file="assets/cardassian.gif")
Falcon = Ship(canvas, falcon_image, 200, 600)

# enterprise
enterprise_image = PhotoImage(file="assets/enterprise.gif")
Enterprise = Ship(canvas, enterprise_image, -1000, 10)

# main game loop
phaserCount = 0
explosionCount = 0
score = 0
while True:
        
    # fly the Falcon NE
    Falcon.fly_NE()
    
    # get falcon position
    pos = Falcon.canvas.coords(Falcon.id)
    # wrap the falcon if it goes off the edge of the canvas
    if pos[0] > 900:
        Falcon.canvas.move(Falcon.id, -750, 750)
    
    
    # handle explosion animation
    if explosionCount == 3:
        ExplosionImage = PhotoImage(file="assets/explosion-med.gif")
        Explosion = Ship(canvas, ExplosionImage, pos[0] - 150, pos[1] - 150)
        explosionCount -= 1
    elif explosionCount == 2:
        ExplosionImage = PhotoImage(file="assets/explosion-lg.gif")
        Explosion = Ship(canvas, ExplosionImage, pos[0] - 300, pos[1] - 300)
        explosionCount -= 1
    elif explosionCount == 1:
        canvas.delete(Explosion.id)
        explosionCount -= 1
        # Add to the score
        score += 5
        scoreText = "Score: " + str(score)
        try:
            canvas.delete(textID)
        except:
            pass
        textID = canvas.create_text(100, 600, text=scoreText, fill="white", font=("Bauhaus 93", 24))
    
    
    # fly the Enterprise E
    Enterprise.fly_E()
    
    # get enterprise position
    pos = Enterprise.canvas.coords(Enterprise.id)
    # wrap the enterprise if it goes off the edge of the canvas
    if pos[0] > 900:
        Enterprise.canvas.move(Enterprise.id, -1900, 0)
    
    # get enterprise position
    pos = Enterprise.canvas.coords(Enterprise.id)
    try:
        Enterprise.canvas.delete(lineID)
    except:
        pass
    if phaserCount > 0:
        phaserStartX = pos[0] + 774
        phaserStartY = pos[1] + 183
        lineID = Enterprise.firePhaser(phaserStartX, phaserStartY, "SE")
        phaserCount -= 2
        
        # if we detect a hit, display explosion animation
        falconPos = Falcon.canvas.coords(Falcon.id)
        falconCenterX = falconPos[0] + 75
        falconCenterY = falconPos[1] + 50
        
        # calculate the slope of the line in relation to the falcon
        diffX = falconCenterX - phaserStartX
        diffY = falconCenterY - phaserStartY
        
        # avoid divide by zero
        if diffY != 0:
            slope = diffX/diffY
            if slope > 0.3 and slope < 2.3:
                explosionCount = 3
                ExplosionImage = PhotoImage(file="assets/explosion-sm.gif")
                Explosion = Ship(canvas, ExplosionImage, falconPos[0], falconPos[1])
        
        
    # print("PhaserCount: %s" % phaserCount)
    
    #tk refresh screen
    tk.update_idletasks()
    tk.update()
    
    # frame rate
    time.sleep(0.005)
    
```
